## 5. Exploratory Data Analysis (EDA)

```python
# Displaying data summary
hit_stats.info()
```

```python
# Displaying Statistical Summaries
hit_stats.describe()
```

```python
# Analyzing Categorical Values
hit_stats.value_counts()
```

```python
# Finding Missing Values
hit_stats.isna().sum()
```

```python
# Dropping missing Values
hit_stats.dropna(inplace=True)
```

```python
# Group by postion and games to see total games played by postion
pos_games = hit_stats.groupby('position')['Games'].sum()
pos_games
```

```python
# Group by postion and average games played to see average games played by postion
pos_games_avg = hit_stats.groupby('position')['Games'].mean()
pos_games_avg
```

```python
# Group by postion and runs to see total runs scored by postion
pos_runs = hit_stats.groupby('position')['Runs'].sum()
pos_runs
```

```python
# Group by postion and average runs scored to see average runs scored by postion
pos_runs_avg = hit_stats.groupby('position')['Runs'].mean()
pos_runs_avg
```

## 6. Data Visualization

```python
# Importing necessary add-ons to display graphs and charts
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Graphing a histogram to show distribution of games played
plt.figure(figsize=(8, 6))
sns.histplot(hit_stats['Games'], kde=True)
plt.title('Distribution of Games Played')
plt.xlabel('Games Played')
plt.ylabel('Count')
sns.despine()
plt.show()
```

This histogram shows the distribution of games played between all players. This distirbution shows us the average amount of games most players play in their careers which is roughly 750 career games played.

```python
# Graphing a barplot to show average games played by position
plt.figure(figsize=(10, 6))
sns.barplot(x=pos_games_avg.index, y=pos_games_avg.values)
plt.title('Average Games Played per Position')
plt.xlabel('Position')
plt.ylabel('Average Games Played')
sns.despine()
plt.show()
```

In this bar plot we are able to see the average amount of games played by position. Here we can see that pitchers played far less from hitters. Most pitchers only play once a week to conserve their arm. Additionally, the DH position was added in 1973 for the American League and in 2022 for the National League. This creates an issue where Designated Hitters (DH) account for only a small portion of games all time. This can be seen more prominent in later graphs. However, according to this chart Designated Hitters(DH) and Shorstops(SS) play on average 100+ more games than other positions.

```python
# Graphing a piechart to show average games played by position
plt.figure(figsize=(10, 6))
plt.pie(pos_games_avg, labels=pos_games_avg.index, autopct='%1.1f%%', startangle=90)
plt.title('Average Games Played by Position')
plt.axis('equal')
sns.despine()
plt.show()
```

This pie chart shows the average amount of games played by position in a different way. Here we see just the exact percentage of the average games played by position. This pie chart supoports the bar chart from earlier stating that Designated Hitters(DH) and Short Stops(SS) on average play more ganes in their careers compared to other positions.

```python
# Graphing a piechart to show total games played by position
plt.figure(figsize=(10, 6))
plt.pie(pos_games, labels=pos_games.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Games Played by Position')
plt.axis('equal')
sns.despine()
plt.show()
```

Here we can see the small amount of games pitchers and Designated Hitters account for. Additionally we can see the Outfield(OF) position also accounts for a small amount of games. As explained earlier pitchers only play once a week and starting back in 1973, pitchers were not really hitting due to the addition of the Designated Hitter in the American League. Additionally, designated hitters were recently added to the National League only in 2022. However why does Outfielders account for such little amount of games? Outfielders are players who slot in anywhere in right, center, or left field. Due to them constantly moving within the outfield they are given the poistion name Outfield and play anywhere in the out field. This makes this position very rare as most outfielders only stick in one spot. This reason is why the outfield position has such little amount of games played in the grand scheme of MLB history.

```python
# Graphing a barplot to show total games played by position
plt.figure(figsize=(10, 6))
sns.barplot(x=pos_games.index, y=pos_games.values)
plt.title('Total Games Played per Position')
plt.xlabel('Position')
plt.ylabel('Total Games Played')
sns.despine()
plt.show()
```

This bar chart shows the total amount of games played by position. Here we can see that 1B or 1st baseman have played the most games all time. In this chart we can see more clearly the amount of games OF's, P's, and DH's amounted for.

```python
# Graphing a barplot to show average runs scored per game by position
plt.figure(figsize=(10, 6))
runs_per_game = pos_runs / pos_games
sns.barplot(x=runs_per_game.index, y=runs_per_game.values)
plt.title('Average Runs per Game by Position')
plt.xlabel('Position')
plt.ylabel('Average Runs per Game')
sns.despine()
plt.show()
```

This bar plot shows us the average amount of runs scored per game by position. Here we can see that CF's provide more runs on a per game basis than other positions. Additionally, besides pitchers who account for a small amount of games, catchers score the least amount of runs on a per game basis.

```python
# Graphing a barplot to show total runs scored by position
plt.figure(figsize=(10, 6))
sns.barplot(x=pos_runs.index, y=pos_runs.values)
plt.title('Total Runs Scored per Position')
plt.xlabel('Position')
plt.ylabel('Total Runs Scored')
sns.despine()
plt.show()
```

This graph shows the total amount of runs scored by each position. Here we can see that 1st basemen have scored the most runs all time. Earlier we also saw that 1st basemen have played more games than any other position all time. There could possibly be a correlation with runs scored and games played. We can also see how outfielders, pitchers, and deisgnated hitters account for a small amount of runs compared to the other positions.

```python
# Graphing a boxplot to show runs scored by postion and any outliers within each postion
plt.figure(figsize=(10, 6))
sns.boxplot(x='position', y='Runs', data=hit_stats)
plt.title('Runs Scored by Position')
plt.xlabel('Position')
plt.ylabel('Runs Scored')
sns.despine()
plt.show()
```

This box chart shows the difference between runs scored and postion. We can see some outliers within certain positions signifying elite players who went above and beyond.

```python
# Graphing a Scatter PLot to show the rleationship between games played and runs scored
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Games', y='Runs', hue='position', data=hit_stats)
plt.title('Relationship between Games Played and Runs Scored')
plt.xlabel('Games Played')
plt.ylabel('Runs Scored')
sns.despine()
plt.show()
```

This scatter plot shows and proves the positve correlation between games played and runs scored. We can see as players play more games they are able to score more runs.
